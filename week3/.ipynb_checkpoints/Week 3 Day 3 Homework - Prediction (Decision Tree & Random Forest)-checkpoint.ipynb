## 1. Decision Tree

* 사전학습: 도메인 지식은 분석에 매우 중요하므로 사전학습을 실시하는 것이 좋다. 가장 좋은 방향은 `데이터 전문가와 도메인 전문가가 협업하여 최적의 모델을 찾아 문제를 해결`하는 것이다. 본 과제에서는 인터넷 검색을 통하여 콘크리트 강도에 영향을 미치는 요인들을 사전 조사하는 방식을 취한다.


* 콘크리트 강도에 영향을 미치는 요인

#### Step 0 : Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

from sklearn.tree import export_graphviz
import graphviz

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

#### Step 1 : Read data using Pandas dataframe

```python
df_raw = pd.read_csv("Concrete.csv")

df_raw.head(5)
```

#### Step 1-1: Missing value & Outlier Detection (Optional)

```python
df_raw.isnull().sum()
```

```python
df_raw.isna().sum()
```

#### Step 2: Data Composition

```python
df_raw_y = df_raw["CONCRETE_COMPRESSIVE_STRENGTH"]
df_raw_x = df_raw.drop("CONCRETE_COMPRESSIVE_STRENGTH", axis = 1, inplace = False)
```

```python
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y, \
                                                                test_size = 0.3, random_state = 1234)
```

```python
df_train_x, df_valid_x, df_train_y, df_valid_y = train_test_split(df_train_x, df_train_y, \
                                                                test_size = 3/7, random_state = 1234)
```

```python
print ("Train data X size\t: {}".format(df_train_x.shape))
print ("Train data Y size\t: {}".format(df_train_y.shape))
print ("Train data X size\t: {}".format(df_valid_x.shape))
print ("Train data Y size\t: {}".format(df_valid_y.shape))
print ("Test data X size\t: {}".format(df_test_x.shape))
print ("Test data Y size\t: {}".format(df_test_y.shape))
```

```python
tree_uncustomized = DecisionTreeRegressor(random_state = 1234)
tree_uncustomized.fit(df_train_x, df_train_y)

print("Score on training set:\t%.3f" %(tree_uncustomized.score(df_train_x, df_train_y)))
print("Score on valid set:\t%.3f" %(tree_uncustomized.score(df_valid_x, df_valid_y)))
print("Score on test set:\t%.3f" %(tree_uncustomized.score(df_test_x, df_test_y)))
```


##### 과대적합
- 파라미터를 조정하지 않은 결정트리를 사용하여 트레이닝 시킨 결과 100% 의 시멘트 강도 예측이 가능한 모델이 만들어졌고, 이 모델을 테스트 세트를 통해 테스트 한 결과 69.9%의 예측치를 보였다. `트레이닝 모델에서는 높은 결과를 보였지만 트레이닝-테스트 세트간 차이가 30% 로 트레이닝 데이터셋에 모델을 최적화한 과적합이 의심된다.` 파라미터 조정을 하여 노드들을 가지치기 하여 과적합을 방지하고 테스트 세트에서도 높은 예측치를 보이는 파라미터를 선정하여야 한다.

##### 가지치기 방식

- `잎사귀 노드 최소 자료 수`: 분리된 노드의 최소자료수치를 설정해 수치 이상의 노드만을 선정한다.
- `분리 노드의 최소 자료 수`: 분리가 되기 위한 부모노드(상위 노드)의 최소 자료수를 설정한다. 넘지 못하는 노드는 분리하지 않는다.
- `최대 깊이`: 최대 깊이의 제한을 두어 이상의 깊이는 노드를 생성하지 않는다.


#### Step 3 : Generating Model & Tunning Parameter
#### Step 3-1 : Tunning Leaf Size

```python
train_score = []; test_score = []
para_leaf = [n_leaf * 1 for n_leaf in range (1, 20)]

for v_min_samples_leaf in para_leaf:
    tree = DecisionTreeRegressor(random_state=1,min_samples_leaf=v_min_samples_leaf)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x, df_train_y))
    test_score.append(tree.score(df_test_x, df_test_y))

df_score_leaf = pd.DataFrame()
df_score_leaf["leaf_size"] = para_leaf
df_score_leaf["TrainScore"] = train_score
df_score_leaf["TestScore"] = test_score
```

```python
df_score_leaf.round(3)
```

```python
plt.plot(para_leaf, train_score, linestyle = "-", label = "Train Score")
plt.plot(para_leaf, test_score, linestyle = "--", label = "Test Score")
plt.legend()
```

#### Step 3-2 : Tunning Minimum Split Size

```python
train_score = []; test_score = []
para_split = [n_leaf * 2 for n_leaf in range (2, 20)]

for v_min_samples_split in para_split:
    tree = DecisionTreeRegressor(random_state=1234,min_samples_leaf=10,
                                min_samples_split=v_min_samples_split)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x, df_train_y))
    test_score.append(tree.score(df_test_x, df_test_y))

df_score_split = pd.DataFrame()
df_score_split["MinSamplesLeaf"] = para_split
df_score_split["TrainScore"] = train_score
df_score_split["TestScore"] = test_score
df_score_split.round(3)
```

```python
df_score_split_diff = pd.DataFrame()
df_score_split_diff["MinSamplesLeaf"] = para_split
df_score_split_diff["TrainScore"] = train_score
df_score_split_diff["TestScore"] = test_score
df_score_split_diff["Diff"] = df_score_split["TrainScore"]-df_score_split["TestScore"]
```

```python
df_score_split_diff.sort_values("Diff").round(3)
```

```python
plt.plot(para_split, train_score, linestyle = "-", label = "Train Score")
plt.plot(para_split, test_score, linestyle = "--", label = "Test Score")
plt.legend()
```

#### Step 3-3 : Tunning Depth

```python
train_score = []; test_score = []
para_depth = [depth for depth in range(1, 11)]

for v_max_depth in para_depth:
    tree = DecisionTreeRegressor(random_state=1234,min_samples_leaf=10,
                                min_samples_split=20, max_depth = v_max_depth)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x, df_train_y))
    test_score.append(tree.score(df_test_x, df_test_y))

df_score_depth = pd.DataFrame()
df_score_depth["Depth"] = para_depth
df_score_depth["TrainScore"] = train_score
df_score_depth["TestScore"] = test_score
df_score_depth.round(3)
```

```python
plt.plot(para_depth, train_score, linestyle = "-", label = "Train Score")
plt.plot(para_depth, test_score, linestyle = "--", label = "Test Score")
plt.legend()
```

### 해석 및 결론

- 최종 모델 결과에서도 알 수 있듯 트레이닝한 모델은 트레이닝 세트와 Validation Set 세트의 예측치의 차이가 85.5% 와 73.1% 로 약 12% 정도 차이나는 걸 알 수 있다. 게다가 성능을 높이기 위하여 파라미터들을 작은 숫자로 주다 보다 그래프가 빽빽하고 어느정도 과대적합 된다. 이를 알 수 있는 것은 파라미터 변경은 테스트 세트나 검증 세트의 예측치 점수에는 큰 영향을 미치지 않는데 트레이닝 세트는 파라미터 점수에 큰 영향을 받는다. 

- 따라서 변수: 시멘트, 고로슬래그, 석탄재, 물, 굵은 골재, 잔골재, 연식을 변수로 결정 트리 예측 모델을 사용할 경우 `20% 내외 구간에서 예측이 틀릴 가능성이 있다.` 콘크리트는 건물에 들어가는 재료이자, 강도는 건물을 지지하는 중요한 요소이다. `강도 예측이 빗나가면 건물의 균열이 생기거나 극단적인 경우 건물 붕괴로 인명사고 까지 이어질 수 있으므로 해당 모델의 현업에서의 사용은 불가능 할 것`이다. 좀 다 나은 파라미터로 튜닝하기 위해서 그리드 서치 알고리즘을 사용할 수 있으나 다른 모델 사용도 고려해 보아야 한다.

- 그리드 서치를 사용하였을 때의 베스트 파라미터를 설정한 결과도 트레이닝 세트에서는 좋은 결과를 보이지만 Valid/Test Set에서 30% 차이를 보이므로 과대적합이며 콘크리트 강도 예측에 사용할 수 없다.

- 중요도 분석 결과 시멘트와 물의 비율이 많은 영향을 미치고 연식 역시 영향을 많이 미친다. 하지만 변수 이름으로만으로는 자세한 내용을 알기 어려움으로 도메인 전문가의 도움이 필요할 것 같다.

## 2. Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
train_score = []; test_score=[]

para_n_tree = [n_tree * 10 for n_tree in range (1,11)]
for v_n_estimators in para_n_tree:
    rf = RandomForestRegressor(n_estimators = v_n_estimators, random_state = 4)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_test_x, df_test_y))

df_score_n = pd.DataFrame()
df_score_n["n_estimator"] = para_n_tree
df_score_n["TrainScore"] = train_score
df_score_n["TestScore"] = test_score

df_score_n.round(3)
```

```python
plt.plot(para_n_tree, train_score, linestyle = "-", label = "Train Score")
plt.plot(para_n_tree, test_score, linestyle = "--", label = "Test Score")
plt.ylabel("score"); plt.xlabel("n_estimators")
plt.legend()
```

```python
train_score = []; test_score=[]

para_leaf = [n_leaf * 1 for n_leaf in range(1, 21)]

for v_min_samples_leaf in para_leaf:
    rf = RandomForestRegressor(n_estimators = 100, min_samples_leaf = v_min_samples_leaf, random_state = 4)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_test_x, df_test_y))
    
df_score_leaf = pd.DataFrame()
df_score_leaf["MinSamplesLeaf"] = para_leaf
df_score_leaf["TrainScore"] = train_score
df_score_leaf["TestScore"] = test_score

df_score_leaf.round(3)
```

```python
plt.plot(para_leaf, train_score, linestyle = "-", label = "Train Score")
plt.plot(para_leaf, test_score, linestyle = "--", label = "Test Score")
plt.ylabel("score"); plt.xlabel("MinSamplesLeaf")
plt.legend()
```

```python
train_score = []; test_score=[]

para_split = [n_split * 2 for n_split in range(2, 21)]

for v_min_samples_split in para_split:
    rf = RandomForestRegressor(n_estimators = 100, min_samples_leaf = 6, min_samples_split = v_min_samples_split, random_state = 4)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x, df_train_y))
    test_score.append(rf.score(df_test_x, df_test_y))
    
df_score_split = pd.DataFrame()
df_score_split["MinSamplesSplit"] = para_split
df_score_split["TrainScore"] = train_score
df_score_split["TestScore"] = test_score
```

### 해석 및 결과


- 따라서 변수: 시멘트, 고로슬래그, 석탄재, 물, 굵은 골재, 잔골재, 연식을 변수로 결정 트리 예측 모델을 사용할 경우 `10% 내외 구간에서 예측이 틀릴 가능성이 있다.` 콘크리트는 건물에 들어가는 재료이자, 강도는 건물을 지지하는 중요한 요소이다. `강도 예측이 빗나가면 건물의 균열이 생기거나 극단적인 경우 건물 붕괴로 인명사고 까지 이어질 수 있으므로 해당 모델의 현업에서의 사용은 불가능 할 것`이다. 좀 다 나은 파라미터로 튜닝하기 위해서 그리드 서치 알고리즘을 사용할 수 있으나 다른 모델 사용도 고려해 보아야 한다.

- `Random Forest 결과` 트레이닝 세트에서 90%의 예측치와 80%의 Validate 세트의 예측치, 테스트 세트에서 77% 예측치를 보여 Decision tree보다 `향상 된 결과를 보였다.` 모델 이상값과 실제값을 빼서 제곱한 MSE/MAE 에서도 Random Forest가 향상된 성능을 보였다. 

- 중요도 분석 결과 시멘트와 물의 비율이 많은 영향을 미치고 연식 역시 영향을 많이 미친다. 하지만 변수 이름으로만으로는 자세한 내용을 알기 어려움으로 도메인 전문가의 도움이 필요할 것 같다.

- 두 방식 모두 파라미터 튜닝을 그래프의 도움이 있을지라도 수작업으로 튜닝 작업을 하여야 하고, `경우의 수가 너무 많아 최적의 값을 찾아내기가 어려웠다.` 그리드 서치의 경우 1000여개 데이터를 돌리는 데에도 컴퓨팅 리소스가 많이 소모되어, 그 이상의 데이터를 처리 할 경우는 효과적이지 않은 방식이라고 생각한다.
