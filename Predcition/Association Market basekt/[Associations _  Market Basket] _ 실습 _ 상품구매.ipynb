```python
import pandas as pd
import numpy as np

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_raw = pd.read_csv("rawdata/상품구매.csv")
df_raw
```

```python
ds_raw = []
ID = list(set(df_raw["ID"]))
ID.sort()

for i in ID:
    temp_list = list(df_raw[df_raw["ID"] == i]["PRODUCT"])
    temp_list.sort()
    ds_raw.append(temp_list)
    
for row in ds_raw:
    print(row)
```

```python
len(ID)
```

```python
enc = TransactionEncoder()
ds_raw_enc = enc.fit_transform(X = ds_raw)
df_asso = pd.DataFrame(ds_raw_enc, columns = enc.columns_)
df_asso.head()
```

```python
min_support = 0.5
df_freq = apriori(df_asso, min_support = 0.5, use_colnames = True)
df_freq.round(3)
```

```python
df_asso_rule = association_rules(df_freq, metric = "confidence", min_threshold = 0.5)
df_asso_rule.round(3)
```
